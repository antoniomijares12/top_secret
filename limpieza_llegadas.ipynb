## Carga, apertura y limpieza de datos:

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import warnings
```

```python
llegadas = pd.read_excel("RAW/DATOS INPUT.xlsx", sheet_name='Aeropuerto', header = 3)  # Tener en cuenta la ruta donde tengas el archivo

llegadas.head(10)
```

```python
# Eliminamos las ultimas columnas, desde la columna 'Total', pues estaban vacías y se cogieron por error:
llegadas = llegadas.loc[:,:'Total']  
llegadas
```

```python
# Empieza en 2005, no necesitamos datos tan antiguos, dejamos desde 2015 (para tener 5 años antes del COVID)
llegadas = llegadas.set_index('Año').loc[2015:]   #Hacemos el set_index para poder referirnos a 2015 con el .loc
llegadas
```

```python
llegadas.duplicated().sum() # Vemos que no hay duplicados
```

```python
llegadas.isnull().sum()  # Estados unidos cuenta con cerca de 1/3 de sus filas nulas, pero como es un origen importante, lo dejamos
```

```python
llegadas.info()  # Tenemos 19 columnas o variables, siendo 1 el mes, otra el total, y 17 los paises de origen.
```

### Pasamos a graficar para ver comportamientos:

#### Estacionalidad de las llegadas:

```python
llegadas_por_mes = llegadas.groupby('Mes').agg({'Total':'sum'})
```

```python
llegadas_por_mes.plot(kind = 'bar', width = 0.7, title = 'Estacionalidad de las llegadas por mes', legend = False);
```

```python
# Vemos como las llegadas suben en los meses de verano, sobre todo Julio, Agosto y Septiembre
```
