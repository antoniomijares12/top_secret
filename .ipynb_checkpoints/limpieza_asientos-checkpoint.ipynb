## Carga, apertura y limpieza de datos:

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import warnings
```

```python
asientos = pd.read_excel("RAW/Datos input.xlsx", sheet_name = 'ASIENTOS OFERTADOS')  # Tener en cuenta la ruta donde tengas el archivo

asientos.head(10)
```

```python
# Eliminamos las 7 primeras filas por estar vacias:
asientos = asientos.iloc[7:]
asientos
```

```python
# Vemos que ahora empieza en agosto de 2008, no necesitamos datos tan antiguos y preferimos tener aÃ±os completos asi que eliminamos hasta 2009:
asientos = asientos.iloc[5:]
asientos
```

```python
asientos.info()
```

### Contamos con 192 filas sin nulos y 54 columnas, de las cuales:
###    - La primera es formato fecha
###    - La segunda es un formato objeto repitiendo la fecha (la eliminamos)
###    - La tercera es el total de asientos reservados por fecha
###    - Las demas, el total de asientos por pais

```python
asientos = asientos.drop(columns = "Fecha 2")
asientos
```

```python
(asientos == 0).sum()  # Localizamos las columnas con muchos registros en 0
```

```python
# Eliminamos los paises con mas de 15 registros siendo 0, pues faltan muchos datos y nos interesan los principales emisores:
columnas_quedan = asientos.columns[(asientos == 0).sum() < 15]
asientos = asientos[columnas_quedan]
asientos
```

```python
asientos.duplicated().sum() # Vemos que no hay duplicados
```

```python
asientos = asientos.reset_index().drop(columns = "index")  # Reseteamos el indice por claridad, al haber borrado filas
```

```python
asientos
```

### Pasamos a graficar para ver comportamientos:

#### Estacionalidad de la oferta:

```python
asientos['num_mes'] = asientos['Fecha'].dt.month # Creamos la columna mes para poder agrupar por mes
asientos
```

```python
asientos_por_mes = asientos.groupby('num_mes').agg({'Total':'sum'})
```

```python
asientos_por_mes.plot(kind = 'bar', width = 0.7, title = 'Estacionalidad de los asientos fletados por mes', legend = False);
```

```python
# Vemos como la oferta de asientos sube en los meses de verano, sobre todo Julio, Agosto y Septiembre
```

#### Evolucion en el tiempo

```python
asientos['year'] = asientos['Fecha'].dt.year
```

```python
asientos_por_aÃ±o = asientos.groupby('year').agg({'Total':'sum'})
```

```python
asientos_por_aÃ±o.plot(kind = 'line', title = 'Evolucion de oferta de asientos por aÃ±o', legend = False);
```

```python
# Observamos que la evolucion es positiva: cada vez mas asientos.
# Se ve tambiÃ©n el gran impacto del Covid en 2020, seguido de la rapida recuperaciÃ³n
```

#### Por Ãºltimo, vemos la cantidad de asientos por paÃ­s, por identificar los paises clave:

```python
# DataFrame vacÃ­o para almacenar los resultados
asientos_pais = pd.DataFrame(columns=['Pais', 'Suma'])

# Iterar sobre cada columna y calcular la suma
for columna in asientos.columns:
    if columna not in ['Fecha', 'year', 'num_mes', 'Total']:
        suma = asientos[columna].sum()
        asientos_pais = pd.concat([asientos_pais, pd.DataFrame({'Pais': [columna], 'Suma': [suma]})], ignore_index=True)

asientos_pais = asientos_pais.set_index('Pais')
asientos_pais
```

```python
asientos_pais.plot(kind = 'bar')
```
